# **Importando e tratando o arquivo .csv**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# from google.colab import files
# uploaded = files.upload()
# df = pd.read_csv("Table 3-1.csv") #Fazer importação do Github
df = pd.read_csv("https://raw.githubusercontent.com/AldoSilva07/Grupo-4---M-dulo-de-Python/refs/heads/main/Table%203-1.csv") #Fazer importação do Github

# Tratamento dos dados
df_clean_lines = df.dropna(thresh=len(df.columns) - 2) # Remove linhas com mais de 2 valores NaN
df_clean = df_clean_lines.drop('Table 3-1:  CO2, CH4, and N2O Emissions from Energy (MMT CO2 Eq.)', axis=1) #Remove coluna vazia
df_clean.columns = df_clean.iloc[0]  # Define os nomes das colunas como sendo a primeira linha
df_clean = df_clean[1:].reset_index(drop=True)  # Remover a linha usada para os nomes e resetar o índice
df_clean.index = df_clean.iloc[:, 0] # Define os nomes das linhas como sendo a primeira coluna
df_clean = df_clean.iloc[:, 1:]  # Remove a primeira coluna
df_clean = df_clean.drop(index="Incineration of Waste", errors='ignore')

# Remover espaços extras e converter para float
df_clean = df_clean.apply(lambda x: x.str.replace(',', '').str.strip() if x.dtype == "object" else x)
df_clean = df_clean.apply(pd.to_numeric, errors='coerce')  # Converte para numérico


display(df_clean)
#print(df_clean.dtypes)  # Verifica os tipos das colunas
```

## Emissões de $\text{CO}_2$ por setor

```python
# Dividir por índice de linhas
df_co2 = df_clean.iloc[1:14]  # Da linha 1 à 13
df_ch4 = df_clean.iloc[15:22]  # Da linha 15 à 21
df_n2o = df_clean.iloc[23:26]  # Da linha 23 à 25
df_co2

def adicionar_total(df, force=False):
    if 'Total' in df.columns and not force:
        print("Coluna Total já existe. Use force=True para sobrescrever.")
        return df

    # Criar cópia
    df_copy = df.copy()

    # Adicionar coluna Total
    df_copy.loc[:, 'Total'] = df_copy.iloc[:, 1:].sum(axis=1)

    return df_copy

# Uso
df_co2 = adicionar_total(df_co2)
df_ch4 = adicionar_total(df_ch4)
df_n2o = adicionar_total(df_n2o)

# Gráfico de Emissão total de CO2 por setor
df_co2_total = df_co2["Total"]

# Estilo
plt.style.use('ggplot')
plt.figure(figsize=(12, 7), facecolor='white')

plt.gca().set_facecolor('white')

# Plotagem
df_co2_total.plot(kind='bar',
                  figsize=(12, 7),
                  color='navy',
                  edgecolor='gray',
                  linewidth=0.2,
                  title="Emissões Totais de CO2 por Setor(MMT)")

# Rótulos dos eixos
plt.xlabel("")
plt.ylabel("")

# Rotacionar rótulos para melhor legibilidade
plt.xticks(rotation=45, ha='right', color='black')
plt.yticks([])

# Adicionar valores nas barras
for i, v in enumerate(df_co2_total):
    plt.text(i, v, f'{v:.0f}',
             ha='center',
             va='bottom',
             fontweight='bold')

plt.tight_layout()
plt.show()
```

Importando dados do PIB mundial e filtrando por local e período.



```python
df_gdp = pd.read_csv("https://raw.githubusercontent.com/AldoSilva07/Grupo-4---M-dulo-de-Python/refs/heads/main/API_NY.GDP.MKTP.KD.ZG_DS2_en_csv_v2_76269.csv", sep=',', skiprows=4) #Fazer importação do Github


# Filtrar a linha referente aos EUA
df_usa = df_gdp[df_gdp["Country Code"] == "USA"]

# Selecionar apenas as colunas dos anos entre 1980 e 2014
years = [str(year) for year in range(1990, 2015)]
df_usa_selected = df_usa[years]

# Corrige o rótulo dos dados
df_usa_selected.columns.values[0] = "PIB"

# Selecionar apenas as colunas dos anos entre 1999 e 2023
years = [str(year) for year in range(1999, 2024)]
df_usa_selected_future = df_usa[years]

# Corrige o rótulo dos dados
df_usa_selected_future.columns.values[0] = "PIB"
teste_predicao_pib = df_usa_selected_future.values

# Exibir os dados filtrados
# df_usa_selected
treinos_predicao_pib = df_usa_selected.values

treinos_predicao_co2 = df_clean.values[0]
treinos_predicao_co2 = treinos_predicao_co2.reshape(1, -1)

y_test = np.zeros(25)

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
import numpy as np


# treino AI
lin_reg = LinearRegression()
lin_reg.fit(treinos_predicao_pib, treinos_predicao_co2)

rf_reg = RandomForestRegressor()
rf_reg.fit(treinos_predicao_pib, treinos_predicao_co2)

# teste AI

test_pred_lin = lin_reg.predict(teste_predicao_pib)
test_pred_rf = rf_reg.predict(teste_predicao_pib)

test_pred_lin = test_pred_lin.flatten()
test_pred_rf = test_pred_rf.flatten()

r2_lin = metrics.r2_score(y_test, test_pred_lin)
mse_lin = metrics.mean_squared_error(y_test, test_pred_lin)
print(f"R² da Regressão Linear: {r2_lin}")
print(f"MSE da Regressão Linear: {mse_lin}")
r2_rf= metrics.r2_score(y_test, test_pred_rf)
mse_rf = metrics.mean_squared_error(y_test, test_pred_rf)
print(f"R² do Random Forest: {r2_rf}")
print(f"MSE do Random Forest: {mse_rf}")


df_resultado = pd.DataFrame()
# df_resultado.index = x_test
# df_resultado['y_teste'] = y_test
df_resultado = test_pred_rf
#df_resultado['y_previsao_lin'] = test_pred_lin
# display(df_resultado)
# df_resultado = df_resultado.reset_index(drop=True)
plt.figure(figsize=(15, 5))
ax = sns.lineplot(data=df_resultado[15:-1])
ax.set(xlabel='Ano')
ax.set_xticklabels(list(range(2014,2024)))
plt.show()
# display(df_resultado)
```

## Comparação de emissões por fonte

```python
# Remover as fontes "CH4", "CO2", "N2O" e "Total" da análise
df_clean_filtered = df_clean.drop(['CH4', 'CO2', 'N2O', 'Total'], axis=0, errors='ignore')

# Escolher os anos específicos para a comparação
anos = ['1990', '2014']

# Plotar o gráfico de barras lado a lado
plt.figure(figsize=(16, 8))  # Aumentar o tamanho do gráfico para melhor visualização

# Definindo a largura das barras para que fiquem lado a lado, de modo a serem proporcionalmente maiores
bar_width = 0.4  # Aumentando a largura das barras para 0.4

# Posições das barras para os dois anos
r1 = range(len(df_clean_filtered))  # Posições para 1990
r2 = [x + bar_width for x in r1]  # Posições para 2014

# Plotar as barras para 1990
plt.bar(r1, df_clean_filtered['1990'], color='skyblue', width=bar_width, label='1990')

# Plotar as barras para 2014
plt.bar(r2, df_clean_filtered['2014'], color='salmon', width=bar_width, label='2014')

# Título e rótulos
plt.title('Comparação de Emissões por Fonte (1990 vs 2014)', fontsize=16, pad=50)  # Ajuste do título para subir 5cm
plt.xlabel('Fonte de Emissões', fontsize=12)
plt.ylabel('Emissões (MMT CO2 Eq.)', fontsize=12)

# Adicionar rótulos nas barras
for i, v in enumerate(df_clean_filtered['1990']):
    # Colocar os números na vertical para o ano de 1990
    plt.text(i, v + 50, f'{v:.2f}', ha='center', fontsize=10, rotation=90)

for i, v in enumerate(df_clean_filtered['2014']):
    # Colocar os números na vertical para o ano de 2014
    plt.text(i + bar_width, v + 50, f'{v:.2f}', ha='center', fontsize=10, rotation=90)

# Ajustar os rótulos do eixo X
plt.xticks([r + bar_width / 2 for r in r1], df_clean_filtered.index, rotation=45, ha='right', fontsize=10)

# Ajustar o eixo Y para ter uma distância de 500 em 500
plt.yticks(range(0, int(df_clean_filtered.max().max()) + 500, 500))

# Adicionar a legenda
plt.legend(title="Ano")

# Remover a borda superior
plt.gca().spines['top'].set_visible(False)

# Ajustar o layout para garantir que tudo se encaixe
plt.tight_layout()

# Exibir o gráfico
plt.show()
```
